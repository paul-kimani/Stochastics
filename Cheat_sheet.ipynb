## 1. Finite Case: Coin Tosses

* **Sample space**:

  $$
  \Omega = \{H, T\}
  $$

  (Heads or Tails)

* **σ-algebra (set of events)**:

  $$
  s^\Omega = \{\emptyset, \{H\}, \{T\}, \{H, T\}\}
  $$

  It contains: impossible event, singletons, and the whole space.

* **Probability measure** $P$:
  Suppose a fair coin:

  $$
  P(\{H\}) = 0.5, \quad P(\{T\}) = 0.5.
  $$

  Then automatically,

  $$
  P(\emptyset) = 0, \quad P(\{H, T\}) = 1.
  $$

✅ This gives us a probability space:

$$
(\Omega, s^\Omega, P).
$$

## 2. Continuous Case: Random variable on the real line

* **Sample space**:

  $$
  \Omega = \mathbb{R} \quad \text{(all possible real outcomes, e.g. lifespans, returns, etc.)}
  $$

* **σ-algebra**:
  Here we cannot take “all subsets of $\mathbb{R}$” (too wild), so we take the **Borel σ-algebra**:

  * smallest σ-algebra containing all open intervals $(a, b)$.
  * This gives us events like $[0,1]$, $(-\infty, 3)$, unions of intervals, etc.

* **Probability measure** $P$:
  For example, suppose $X \sim N(0,1)$ (standard normal distribution). Then:

  $$
  P([0,1]) = \int_0^1 \frac{1}{\sqrt{2\pi}} e^{-x^2/2} dx.
  $$

  More generally:

  $$
  P(A) = \int_A f(x)\, dx,
  $$

  where $f$ is the probability density function (pdf).

✅ Probability space:

$$
(\mathbb{R}, \text{Borel σ-algebra}, P).
$$

---

### Key contrast:

* In finite cases: σ-algebra is just all subsets of Ω.
* In continuous cases: σ-algebra must be carefully defined (usually Borel), and probability is given by integrating a density.

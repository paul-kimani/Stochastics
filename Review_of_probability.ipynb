## 1. Measure
###  **Measure in General**

In mathematics, a **measure** is a systematic way of assigning a “size” or “weight” to subsets of a given set. Think of it as a generalization of:

* **Length** on a line (intervals in $\mathbb{R}$)
* **Area** on a plane (regions in $\mathbb{R}^2$)
* **Volume** in space (regions in $\mathbb{R}^3$)

But it goes beyond geometry: a measure doesn’t have to be about physical size; it could represent probability, mass, or even abstract notions like "likelihood."

Formally, if $(\Omega, \mathcal{F})$ is a measurable space:

* $\Omega$ is the sample space (all possible outcomes),
* $\mathcal{F}$ is a σ-algebra (collection of “nice” subsets of $\Omega$),

then a **measure** is a function $\mu: \mathcal{F} \to [0, \infty]$ that satisfies:

1. **Non-negativity:** $\mu(A) \geq 0$ for all $A \in \mathcal{F}$.
2. **Null empty set:** $\mu(\emptyset) = 0$.
3. **Countable additivity (σ-additivity):** If $A_1, A_2, \dots$ are disjoint sets in $\mathcal{F}$, then

   $$
   \mu\left(\bigcup_{i=1}^{\infty} A_i\right) = \sum_{i=1}^{\infty} \mu(A_i).
   $$

That’s the formal backbone.

---

### **Measure in Stochastics**

In **stochastics** (the study of randomness), the central measure we care about is the **probability measure**.

* Here, $\Omega$ = sample space (all possible outcomes, e.g., coin flips, stock price paths).
* $\mathcal{F}$ = σ-algebra of events (the collection of all events we can talk about).
* A **probability measure** $P: \mathcal{F} \to [0,1]$ assigns a “probability weight” to each event.

It satisfies the same rules as a measure **plus one extra condition**:

$$
P(\Omega) = 1
$$

because the total probability must be 1 (something in the sample space must happen).

So, in stochastics:

* A **measure** is the tool we use to assign probabilities, expectations, and distributions to events and random variables.
* Every probability distribution (Normal, Poisson, Binomial, etc.) is just a probability measure on some measurable space.

* In General, A **Measure** is a set function that assigns consistent magnitude to measurable subsets of a measurable space.

* $\Omega$ - is the set of all possible outcomes

---

###  **Applications**

* **Probability space:** $(\Omega, \mathcal{F}, P)$ is the foundation of all probability theory.
* **Expectation:** Defined as an integral with respect to a probability measure:

  $$
  \mathbb{E}[X] = \int_\Omega X(\omega) \, dP(\omega).
  $$
* **Change of measure:** In financial mathematics, we often switch from the "real-world" measure $P$ to the "risk-neutral" measure $Q$ when pricing derivatives.

---

###  **Simple Example**

* Toss a fair coin:

  * $\Omega = \{\text{Heads}, \text{Tails}\}$.
  * $\mathcal{F} = \{\emptyset, \{H\}, \{T\}, \{H,T\}\}$.
  * Define measure $P$:

    * $P(\{H\}) = 0.5$, $P(\{T\}) = 0.5$, $P(\emptyset) = 0$, $P(\{H,T\}) = 1$.

This is a **probability measure** on a very simple space.


## 2. Measurable space

###  **What is a measurable space?**

A **measurable space** is a pair

$$
(\Omega, \mathcal{F}),
$$

where:

* $\Omega$ is the **sample space** (the set of all possible outcomes).
* $\mathcal{F}$ is a **σ-algebra** (sigma-algebra): a collection of subsets of $\Omega$ that are considered **measurable events**.

So a measurable space is basically a set plus a structure that tells us which subsets are “legal” to assign measures (probabilities) to.

We could also say that it consists of a set $\Omega$ and a $\sigma$-algebra

---

### **Why do we need a σ-algebra?**

Not every subset of $\Omega$ is “nice enough” to assign a measure to. Some sets are too pathological (like certain subsets of the real line).
The σ-algebra ensures:

1. **Closed under complements**: if $A \in \mathcal{F}$, then $\Omega \setminus A \in \mathcal{F}$.
2. **Closed under countable unions**: if $A_1, A_2, \dots \in \mathcal{F}$, then $\bigcup_i A_i \in \mathcal{F}$.
3. By De Morgan’s laws, it’s also closed under countable intersections.

This guarantees consistency when defining measures.

---

### **From measurable space to measure space**

* A **measurable space** $(\Omega, \mathcal{F})$ is just the structure: outcomes + events.
* When we add a measure $\mu$ to it, we get a **measure space**:

  $$
  (\Omega, \mathcal{F}, \mu).
  $$
* If $\mu$ is a probability measure (i.e., $\mu(\Omega) = 1$), then we get a **probability space**:

  $$
  (\Omega, \mathcal{F}, P).
  $$

This triple $(\Omega, \mathcal{F}, P)$ is the **foundation of stochastics**.

---

### **Examples**

#### (a) **Finite sample space**

Coin toss:

* $\Omega = \{H, T\}$.
* $\mathcal{F} = \{\emptyset, \{H\}, \{T\}, \{H,T\}\}$.
  This is the measurable space. If we assign probabilities $P(H)=P(T)=0.5$, we get a probability space.

#### (b) **Continuous sample space**

Random variable on the real line:

* $\Omega = \mathbb{R}$.
* $\mathcal{F} = \mathcal{B}(\mathbb{R})$, the **Borel σ-algebra** (all sets generated from open intervals, e.g. $(a, b]$).
  This ensures we can measure “nice” sets like intervals, unions, intersections.

If we assign measure $P$ as “Normal distribution with mean 0 and variance 1,” then

$$
P((a,b]) = \int_a^b \frac{1}{\sqrt{2\pi}} e^{-x^2/2} dx.
$$

---

**Summary **:
A **measurable space** is the pair $(\Omega, \mathcal{F})$, consisting of a set of outcomes $\Omega$ and a σ-algebra $\mathcal{F}$ of events. It provides the structure needed to define a measure. Adding a measure $\mu$ produces a **measure space**, and if $\mu(\Omega) = 1$, it becomes a **probability space**.

Therefore a **probability space** is the trio $(\Omega, \mathcal{F}, P)$ that model a random experiment. It's the main set-up in which probability theory works.

**NOTE:**
$2^\Omega$ => is all the subsets that can be derived from the sample space.


## 3. Measure space

* **Measurable space** $(\Omega, \mathcal{F})$ → the *stage* (outcomes + events).
  Now we add a **measure** on top of it.

---

###  **Definition**

A **measure space** is a triple

$$
(\Omega, \mathcal{F}, \mu),
$$

where:

* $\Omega$: the sample space (set of all possible outcomes).
* $\mathcal{F}$: a σ-algebra on $\Omega$ (the collection of measurable subsets/events).
* $\mu: \mathcal{F} \to [0, \infty]$: a **measure** that assigns a size, weight, or probability to each event in $\mathcal{F}$.

So basically: a measure space is just a measurable space equipped with a measure.

---

### **Properties**

The measure $\mu$ satisfies:

1. $\mu(\emptyset) = 0$.
2. $\mu(A) \geq 0$ for all $A \in \mathcal{F}$.
3. **Countable additivity (σ-additivity):** if $A_1, A_2, \dots$ are disjoint, then

   $$
   \mu\left(\bigcup_{i=1}^{\infty} A_i\right) = \sum_{i=1}^{\infty} \mu(A_i).
   $$

---

###  **Types of measure spaces**

* **General measure space**: no restriction on total measure. Example: Lebesgue measure on $\mathbb{R}$, where intervals like $[0,1]$ have length 1, but $\mu(\mathbb{R}) = \infty$.
* **Finite measure space**: total measure is finite ($\mu(\Omega)<\infty$).
* **Probability space**: special case where $\mu(\Omega) = 1$. In stochastics, this is the one we usually use.

---

###  **Examples**

#### (a) Counting measure

* $\Omega = \mathbb{N} = \{1,2,3,\dots\}$.
* $\mathcal{F} = 2^{\mathbb{N}}$ (all subsets).
* $\mu(A) = |A|$ if $A$ is finite, and $\infty$ otherwise.

This measure space just “counts” the number of elements.

#### (b) Lebesgue measure

* $\Omega = \mathbb{R}$.
* $\mathcal{F} = \mathcal{L}$ (the Lebesgue σ-algebra).
* $\mu$ = Lebesgue measure (length). So, $\mu([a,b]) = b-a$.

This is the standard way we define “length” rigorously.

#### (c) Probability space

* $\Omega = \{H, T\}$.
* $\mathcal{F} = \{\emptyset, \{H\}, \{T\}, \{H,T\}\}$.
* $\mu = P$, with $P(H)=0.5$, $P(T)=0.5$.

Here, $(\Omega, \mathcal{F}, P)$ is a probability space (total measure = 1).

---

**Summary**

A **measure space** is the triple $(\Omega, \mathcal{F}, \mu)$ consisting of a set, a σ-algebra of subsets, and a measure $\mu$. It generalizes the idea of length, area, volume, and probability. If $\mu(\Omega) = 1$, it is called a **probability space**, the foundation of stochastics.

## 4. Events and Probability

Let $\Omega$ be a sample space.
The collection of events we can measure (denoted $\mathcal{F}$ or here $s^\Omega$) must satisfy the following properties:

1. **Contains the empty set:**

   $$
   \emptyset \in s^\Omega
   $$

   (The “impossible event” must always be measurable.)

2. **Closed under complements:**
   If $A \in s^\Omega$, then its complement is also in $s^\Omega$:

   $$
   A^c = \Omega \setminus A \in s^\Omega.
   $$

3. **Closed under countable unions:**
   If $A_1, A_2, \dots \in s^\Omega$, then

   $$
   \bigcup_{i=1}^{\infty} A_i \in s^\Omega.
   $$

---

### Additional note

* From these three rules, it automatically follows (by De Morgan’s laws) that $s^\Omega$ is also **closed under countable intersections**.
* Such a collection $s^\Omega$ is called a **σ-algebra** (sigma-algebra).

---

### Probability

Once we have this σ-algebra of events, a **probability measure** is a function $P : s^\Omega \to [0,1]$ satisfying:

1. $P(\emptyset) = 0$.
2. $P(\Omega) = 1$.
3. For disjoint sets $A_1, A_2, \dots \in s^\Omega$,

   $$
   P\!\left(\bigcup_{i=1}^{\infty} A_i\right) = \sum_{i=1}^{\infty} P(A_i).
   $$

This makes $(\Omega, s^\Omega, P)$ a **probability space**.



## 5. σ-Algebra (Field of Sets)

Let $\Omega$ be a non-empty set (the sample space).
A **σ-algebra** $\mathcal{F}$ on $\Omega$ is a collection of subsets of $\Omega$ such that:

1. $\emptyset \in \mathcal{F}$.
2. If $A \in \mathcal{F}$, then its complement $A^c = \Omega \setminus A \in \mathcal{F}$.
3. If $A_1, A_2, \dots \in \mathcal{F}$, then

   $$
   \bigcup_{i=1}^{\infty} A_i \in \mathcal{F}.
   $$

By De Morgan’s laws, closure under countable intersections also follows.
Thus, $\mathcal{F}$ defines the collection of **events** on which probabilities can be assigned.

**Example (Coin Toss):**

* Sample space: $\Omega = \{H, T\}$.
* σ-algebra:

  $$
  \mathcal{F} = \{\emptyset, \{H\}, \{T\}, \{H, T\}\}.
  $$

This satisfies all properties (contains ∅, closed under complements, closed under unions).
---

## 6. Probability Measure

A **probability measure** is a function

$$
P : \mathcal{F} \to [0,1]
$$

satisfying:

1. $P(\Omega) = 1$.
2. If $A_1, A_2, \dots$ are pairwise disjoint events in $\mathcal{F}$, then

   $$
   P\left(\bigcup_{i=1}^{\infty} A_i\right) = \sum_{i=1}^{\infty} P(A_i).
   $$

**Example (Dice):**

* $\Omega = \{1,2,3,4,5,6\}$.
* σ-algebra: all subsets of $\Omega$ (since it’s finite).
* Probability measure (fair dice):

  $$
  P(\{k\}) = \tfrac{1}{6}, \quad k=1,\dots,6.
  $$
* Example calculation:

  $$
  P(\text{even}) = P(\{2,4,6\}) = \tfrac{1}{6} + \tfrac{1}{6} + \tfrac{1}{6} = \tfrac{1}{2}.
  $$

---

## 7. Probability Space

The triple

$$
(\Omega, \mathcal{F}, P)
$$

is called a **probability space**, where:

* $\Omega$ = sample space (all possible outcomes),
* $\mathcal{F}$ = σ-algebra of events,
* $P$ = probability measure.

Each $A \in \mathcal{F}$ is called an **event**.
If $P(A) = 1$, the event is said to occur **almost surely (a.s.)**.

---

## 8. Conditional Probability

For events $A, B \in \mathcal{F}$ with $P(B) > 0$, the **conditional probability** of $A$ given $B$ is defined as:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}.
$$

* Intuition: We restrict attention to $B$, then ask what fraction of $B$ also satisfies $A$.
* Diagrammatically: overlap of $A$ and $B$ relative to the size of $B$.

**Example (Dice):**
Let $\Omega = \{1,2,3,4,5,6\}$ with equal probability.

* $A$: outcome is even = $\{2,4,6\}$.
* $B$: outcome > 3 = $\{4,5,6\}$.
* Intersection: $A \cap B = \{4,6\}$.

Now,

$$
P(A \mid B) = \frac{P(\{4,6\})}{P(\{4,5,6\})} = \frac{\tfrac{2}{6}}{\tfrac{3}{6}} = \tfrac{2}{3}.
$$

Interpretation: if you know the roll is >3, the chance it’s even is $\tfrac{2}{3}$.

## 9. Filtration (Information Growth Over Time)

A **filtration** $\{\mathcal{F}_t\}_{t \geq 0}$ is an increasing family of σ-algebras:

$$
\mathcal{F}_0 \subseteq \mathcal{F}_1 \subseteq \cdots
$$

* It represents the flow of information over time.
* $\mathcal{F}_t$ contains all events that are **observable/knowable** up to time $t$.

**Example (Stock Prices):**

* Let $S_t$ = stock price at time $t$.
* $\mathcal{F}_0$: only trivial events (no price observed yet).
* $\mathcal{F}_1$: events based on knowing $S_1$.
* $\mathcal{F}_2$: events based on knowing $S_1, S_2$.
* … and so on.

Interpretation: as time passes, you gain more information about the process.
---

## 10. Martingales (Preview)

A process $\{M_t\}$ is a **martingale** with respect to a filtration $\{\mathcal{F}_t\}$ if:

$$
\mathbb{E}[M_{t+1} \mid \mathcal{F}_t] = M_t.
$$

* Intuition: The **best predictor of tomorrow’s value given today’s information is today’s value itself**.
* Martingales model “fair games.”


**Example (Gambling Game):**

* Flip a fair coin repeatedly.
* Win $+1$ if heads, lose $-1$ if tails.
* Let $M_t$ = total fortune after $t$ flips.

Then:

$$
\mathbb{E}[M_{t+1} \mid \mathcal{F}_t] = M_t.
$$

Because the expected gain on the next flip is zero — it’s a fair game.


---

**Final Structure**:

* **σ-Algebra** → defines measurable events.
* **Probability Measure** → assigns probability consistently.
* **Probability Space** → $(\Omega, \mathcal{F}, P)$.
* **Conditional Probability** → probability of $A$ under knowledge of $B$.
* **Filtration** → evolution of knowledge over time.
* **Martingales** → fair game processes relative to filtrations.
